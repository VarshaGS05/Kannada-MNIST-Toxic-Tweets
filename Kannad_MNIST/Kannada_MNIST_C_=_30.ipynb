```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.load('X_kannada_MNIST_train.npz')['arr_0']
y_train = np.load('y_kannada_MNIST_train.npz')['arr_0']
X_test =  np.load('X_kannada_MNIST_test.npz')['arr_0']
y_test =  np.load('y_kannada_MNIST_test.npz')['arr_0']
```

```python
X_train.shape
```

```python
X_train_reshape = X_train.reshape(X_train.shape[0], 784)
X_test_reshape = X_test.reshape(X_test.shape[0], 784)
```

```python
X_train_reshape
```

```python
X_test_reshape
```

```python
x_train = pd.DataFrame(X_train_reshape)
x_test = pd.DataFrame(X_test_reshape)
```

```python
x_train
```

```python
x_test
```

#**Visualizing Random Inputs**

```python
def plot_random_digit():
  random_index = np.random.randint(0,X_train.shape[0])
  plt.imshow(X_train[random_index], cmap='BuPu_r')
  plt.title(y_train[random_index])
  plt.axis('off')
plt.figure(figsize=[2,2])
plot_random_digit()
```

#**Performing PCA**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=30)
```

```python
pca.fit(x_train)
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
x_train_pca = pd.DataFrame(x_train_pca)
x_test_pca = pd.DataFrame(x_test_pca)
```

```python
x_train_pca
```

```python
x_test_pca
```

#**Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
mm.fit(x_train_pca)
```

```python
x_train_pca_norm = mm.transform(x_train_pca)
x_train_pca_norm = pd.DataFrame(x_train_pca_norm)
x_train_pca_norm
```

```python
x_test_pca_norm = mm.transform(x_test_pca)
x_test_pca_norm = pd.DataFrame(x_test_pca_norm)
x_test_pca_norm
```

```python
x_train_pca_norm.describe()
```

```python
x_test_pca_norm.describe()
```

#**K-NN Classifier**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
params = {'n_neighbors': range(0,10),
          'weights':['uniform', 'distance'],
          'metric':['manhattan','eculidean','cosine']}
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
skf = StratifiedKFold(n_splits = 5)

rcv_knn = RandomizedSearchCV(knn, param_distributions = params, cv = skf, scoring = 'accuracy')
rcv_knn.fit(x_train_pca_norm, y_train)
```

```python
rcv_knn.best_params_
```

```python
rcv_knn.best_score_
```

```python
finalmodel_knn = knn.set_params(**rcv_knn.best_params_)
finalmodel_knn.fit(x_train_pca_norm, y_train)
```

```python
preds_knn = finalmodel_knn.predict(x_test_pca_norm)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Calculating the Evaluation Metrics
precision = precision_score(y_test, preds_knn,average = 'macro')
recall = recall_score(y_test, preds_knn,average = 'macro')
f1 = f1_score(y_test, preds_knn,average = 'macro')
cm_knn = confusion_matrix(y_test, preds_knn)
accuracy = accuracy_score(y_test, preds_knn)


# Computing the ROC curve
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
y_test_bin = label_binarize(y_test,classes=range(10))
predict_proba_knn = finalmodel_knn.predict_proba(x_test_pca_norm)
fpr_knn,tpr_knn,thersholds_knn = roc_curve(y_test_bin.ravel(),predict_proba_knn.ravel())
auc_knn = auc(fpr_knn,tpr_knn)


# Printing the Evalution Metrics Score
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("Accuracy:", accuracy)


# Ploting the Heatmap
sns.heatmap(cm_knn, cmap='coolwarm', annot = True, fmt =".0f")
plt.title("Confusion Matrix")
plt.figure(figsize=(6, 6))
plt.show()


# Plot ROC curve
plt.figure(figsize=(6, 6))
plt.plot(fpr_knn, tpr_knn, label='KNN (AUC = {:.2f})'.format(auc_knn))
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()

```
